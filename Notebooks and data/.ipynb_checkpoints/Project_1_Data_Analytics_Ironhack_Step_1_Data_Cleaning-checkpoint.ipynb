## 1. CLEANING DATASET AMAZON SALES 2020-21

```python
df_sales = pd.read_csv('Amazon Sales FY2020-21.csv')
```

```python
df_sales.head()
```

```python
def clean_sales(df):
    

    # 1.change column names to lower case and spaces for _
    columns = []
    for col in df.columns: 
        if ' ' in col:
            col = col.replace(' ','_')
            columns.append(col.lower())
        else:
            columns.append(col.lower())
            
    df.columns = columns
    
    
    # 2. Change order_date column to datetime format.

    df['order_date'] = pd.to_datetime(df['order_date'])
    df['month_order_date'] = pd.DatetimeIndex(df['order_date']).month
    df['day_week_order_date'] = df['order_date'].dt.day_of_week
    df['year'] = df['order_date'].dt.year
    
    # 3. Drop columns containing repeated or unhelpful information for our analysis


    df = df.drop(columns = ['order_date','month', 'sku','bi_st','sign_in_date','user_name'],axis=1)
    
    return df
```

```python
sales = clean_sales(df_sales)
```

## 2. CLEANING DATASET US CITIES DENSITY 2022

```python
df_popul = pd.read_csv('population-density-v1.csv')
df_popul
```

```python
def clean_population(df):


    # 1.change column names to lower case and spaces for _
    
    columns = []
    for col in df.columns: 
        if ' ' in col:
            col = col.replace(' ','_')
            columns.append(col.lower())
        else:
            columns.append(col.lower())
            
    df.columns = columns
    
    
    # 2. Drop columns with unhelpful information for our analysis

    df = df.drop(['country', 'coordinates', 'cityst'],axis=1)
    
    
    # 3.Replace NaN values in categorical with the mode

    df['st'] = df['st'].fillna('Texas')

    return  df
```

```python
population = clean_population(df_popul)
```

## 3. CLEANING DATASET US AMAZON CENTERS 2022

```python
df_centers = pd.read_csv('Amazon_Centers_USA.csv', sep=';')
df_centers
```

```python
def clean_centers(df):
    

    # 1.change column names to lower case and spaces for _
    
    columns = []
    for col in df.columns: 
        if ' ' in col:
            col = col.replace(' ','_')
            columns.append(col.lower())
        else:
            columns.append(col.lower())
            
    df.columns = columns
    
    
    
    # 2. Change order_date column to datetime format.

    df['date_added'] = pd.to_datetime(df['date_added'])
    df['year'] = df['date_added'].dt.year
    
    
    
    # 3. Filter year to before 2020 (df sales is from 2020-21)
    
    df_filter_centers = df[df['year']< 2020].reset_index(drop=True)
    
    
    
    # 4. Drop columns with dates

    df_filter_centers = df_filter_centers.drop(columns = ['date_added', 'year'])
    df_filter_centers
    
    
    
    
    # 5.Clean zip codes 
    
    df_filter_centers['zip'] = df_filter_centers['zip'].str.split('-', expand=True)[0]
    df_filter_centers['zip'] = df_filter_centers['zip'].str.strip() #to take off all the spaces
    
    code_zip = []
    for i in df_filter_centers['zip']:
        if(len(i)) == 4:
            code_zip.append('0'+i)
        else:
            code_zip.append(i)
            
    df_filter_centers['zip'] = code_zip


        
    # 6.Change capitalization city column

    city_cap = []
    
    for i in df_filter_centers['city']:
        city_cap.append(i.lower().title())
        
    df_filter_centers['city'] = city_cap       
            
    return df_filter_centers
    
```

```python
centers = clean_centers(df_centers)
centers
```
